```python
import requests
from bs4 import BeautifulSoup
```

```python
header = {
    'User-Agent': 'Mozilla/5.0 (hp-desktop; Linux; hpwOS/2.0; U; en-US) AppleWebKit/534.6 (KHTML, like Gecko) wOSBrowser/234.83 Safari/534.6 Desktop/1.0'
    }
baseurl = "https://www.adrodrinternational.com/"
professional_type = ["foundation", "certified", "fellow"]
profilelinks = []

for ptype in professional_type:
  page_url = baseurl+ptype+"-professionals"
  r = requests.get(page_url, headers=header)
  soup = BeautifulSoup(r.content, 'html5lib')
  profilelist = soup.find_all('div', class_="caption-container")

  for item in profilelist:
    for link in item.find_all('a', href=True):
      profilelinks.append(baseurl + link['href'])
```

```python
profile_data = {}
area_ids = ['1603015602', '1366931551']
profmem_ids = ['1354943196', '1291219366', '1196161085']
profexp_ids = ['1443163784', '1656503218', '1283183852']
style_texts = ["Style & Approach", "Mediation Style"]
tags = ['font', 'span']

c = 0
for link in profilelinks:
    r = requests.get(link, headers=header)
    soup = BeautifulSoup(r.content, 'html5lib')
    try:
        name = soup.find('h2', class_='u_1750228315').text.strip()
        profile_data[name]={}

        area = lambda aid: soup.find('div', class_='dmNewParagraph', id=aid)
        area_text = area(area_ids[0]).text if area(area_ids[0]) else area(area_ids[1]).text

        head_fn = lambda aid: soup.find('h2', id=aid)
        profmem_head = head_fn(profmem_ids[0]) if head_fn(profmem_ids[0]) else head_fn(profmem_ids[1]) if head_fn(profmem_ids[1]) else head_fn(profmem_ids[2])
        profmem = profmem_head.find_next('div').text

        profexp_head = soup.find('font', string=lambda text: text and "Professional Experience" in text)
        profexp = profexp_head.find_next('div').text

        style_head_fn = lambda subtext, tag : soup.find(tag, string=lambda text: text and subtext in text)
        style_head = style_head_fn(style_texts[0], tags[0]) if style_head_fn(style_texts[0], tags[0]) else style_head_fn(style_texts[1], tags[0]) if style_head_fn(style_texts[1], tags[0]) else style_head_fn(style_texts[0], tags[1]) if  style_head_fn(style_texts[0], tags[1]) else style_head_fn(style_texts[1], tags[1])
        style = style_head.find_next('div').text

        profile_data[name]["Area of expertise"] = area_text
        profile_data[name]["Professional memberships"] = profmem
        profile_data[name]["Professional experience"] = profexp
        profile_data[name]["Style & Approach"] = style

    except Exception as e:
        print(f"Exception {e} for {link}")
        c+=1
print(c)

# print(len(profile_data))
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(profile_data, orient='index')
df
```

```python
df = df.applymap(lambda text: text.replace("\t", " ").replace("\n", " ").replace("•", " ").replace("\xa0", " ").replace("  ", " ") if isinstance(text, str) else text)

df
```

```python
df.to_csv("mediator_profiles.csv")
```
